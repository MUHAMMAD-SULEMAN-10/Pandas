## Windowing function in pandas

```python
# rolling fun  provide a window calculation like a roll on the data values one by one

# rolling window taking the parameter size = 1 and same data is printing

df4['a'].rolling(window=1).mean()
```

```python
# when window size = 2 and he needs two data like first he take the value 3.0 from a col and 2nd data above its is not availabe so value is NaN

df4['a'].rolling(window=2).mean()
```

```python
# now when the size = 3 then he take three data sizes so when he take first data he needs to take 3 vlue form a col then he need two more vlues but no above so NaN also
# and also when he take 2nd values then he need to take 2 more values but only 1 available so NaN will write.
# AT 5 VALUES above two values are present and calculate the mean()


df4['a'].rolling(window=3).mean()
```

```python
# take a sum by using rolling

df4['a'].rolling(window=3).sum()
```

```python
df4['a'].rolling(window=3).min()
```

```python
df4['a'].rolling(window=3).max()
```

```python
# do cumulative Sum
# sum previous values and give next vlaue one by one ...


df4['a'].cumsum()
```
